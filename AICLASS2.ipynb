```python
import requests

# URL of the text file
text = "https://www.gutenberg.org/files/1524/1524-0.txt"

# send HTTP GET request
response = requests.get(text)

# get the text content
content = response.text

print(content)  # prints the whole text
print("Using Upper case:", content.upper())
print("Using Lower case:", content.lower())
print("Using Title case:", content.title())
print("Using Capitalize:", content.capitalize())
print("Using Swapcase:", content.swapcase())
print("Remove White Space from both end", content.strip())
print("Remove White Space from left end", content.lstrip())
print("Remove White Space from right end", content.rstrip())
print("Replace part of string 'data'",content.replace("data","data science"))
print("Split String by seperator into list", content.split())
print("Join list elements with separator", content.join())

```
